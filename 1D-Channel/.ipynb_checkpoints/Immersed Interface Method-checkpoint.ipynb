```python
import numpy as np
import matplotlib.pyplot as plt
import NSsolver as ns
try:
    plt.style.use("jupyter")
except OSerror:
    print("Using default ploting style")
```

```python
def get_pressure_jump(x):
    #pressure jump is equal to mass a fluid in system
    C_h = (x.max() - x.min())
    return C_h

def kernel(d):
    return np.where(abs(d)<=.5, .75-d**2, np.where(abs(d)<=1.5,.5*(2.25-3*abs(d)+d**2), 0))

def update(x, u, V, X, dx, dt, t):
    
    # get body position
    C_h = get_pressure_jump(x)

    # predict velocity, =0 
    u_star = u + ns.r(u, dx)
    
    # get pressure
    sigma = ns.div(u_star, dx) + C_h*ns.div(kernel((x-X)/X), dx)
    p = ns.solve_pressure(np.ones_like(sigma), sigma, dx)
    
    # correct
    u_n = u_star - dt*ns.grad(p, dx) + dt*C_h*kernel((x-X)/X)
    
    return sigma, p, u_n
```

```python
# plt.plot(x,kernel((x-X)/X));
```

```python
N = 32
x, dx = np.linspace(-1, 1, N, retstep=True)
xs = x +0.5*dx
X = 0.2
V = 1
u0 = np.zeros_like(x)
dt = 1.

sigma, p, u_n =  update(x, u0, V, X, dx, 1, 1)
print("Interface at X: %.2f" % X)
print(r"L inf: %.3e" % np.max(np.abs(u_n -  V)))
plt.plot(xs, u0, '-.k', lw=1, label=r"$u^n$")
plt.plot(xs, sigma*0.1, '-.', lw=1, label=r"div($u^n$)")
plt.plot(xs, p, '--k',  lw=1, label=r"$p^n$")
plt.plot(xs, u_n, '-m', lw=1, label=r"$u^{n+1}$")
plt.ylim(-3,3); plt.xlim(-1,1)
plt.xlabel(r'$x/L$');
plt.legend();
# plt.savefig('Figure_2.png', dpi=900);
```
